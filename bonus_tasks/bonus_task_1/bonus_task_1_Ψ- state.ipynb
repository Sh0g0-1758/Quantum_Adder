let us take |a> = |0000>, |b> = |1111> and entangle the second qubits (i.e a1 and b1). 

```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
from pylatexenc import *
```

```python
qc = QuantumCircuit(13,5)
```

```python
# setting the qubits
qc.x(2)
qc.x(5)
qc.x(8)
qc.x(11)

qc.barrier()
```

```python
# entangling the qubits
qc.h(4)
qc.cx(4,5)
qc.barrier()
```

```python
qc.ccx(1,2,3)
qc.cx(1,2)
qc.ccx(0,2,3)
qc.cx(0,2)

qc.barrier()
```

```python
qc.ccx(4,5,6)
qc.cx(4,5)
qc.ccx(3,5,6)
qc.cx(3,5)

qc.barrier()
```

```python
qc.ccx(7,8,9)
qc.cx(7,8)
qc.ccx(6,8,9)
qc.cx(6,8)

qc.barrier()
```

```python
qc.ccx(10,11,12)
qc.cx(10,11)
qc.ccx(9,11,12)
qc.cx(9,11)

qc.barrier()
```

```python
qc.measure(2,0)
qc.measure(5,1)
qc.measure(8,2)
qc.measure(11,3)
qc.measure(12,4)
```

```python
qc.draw(output='mpl')
```

```python
state_sim = Aer.get_backend("qasm_simulator")
job = execute(qc, state_sim)
counts = job.result().get_counts(qc)

print(counts)
```

```python
plot_histogram(counts)
```

On entangling the two qubits, we get - state. Thus the circuit thinks that 50% of the time, a1=1, b1=0 and 50% of time a1=0, b1=1, thus giving us the sum as 1111
