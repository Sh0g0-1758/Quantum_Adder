let us take |a> = |0000>, |b> = |1110> and entangle the first qubits (i.e a0 and b0).

```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
from pylatexenc import *
```

```python
qc = QuantumCircuit(13,5)
```

```python
# setting the qubits
qc.x(5)
qc.x(8)
qc.x(11)

qc.barrier()
```

```python
# entangling the qubits
qc.h(1)
qc.cx(1,2)
qc.barrier()
```

```python
qc.ccx(1,2,3)
qc.cx(1,2)
qc.ccx(0,2,3)
qc.cx(0,2)

qc.barrier()
```

```python
qc.ccx(4,5,6)
qc.cx(4,5)
qc.ccx(3,5,6)
qc.cx(3,5)

qc.barrier()
```

```python
qc.ccx(7,8,9)
qc.cx(7,8)
qc.ccx(6,8,9)
qc.cx(6,8)

qc.barrier()
```

```python
qc.ccx(10,11,12)
qc.cx(10,11)
qc.ccx(9,11,12)
qc.cx(9,11)

qc.barrier()
```

```python
qc.measure(2,0)
qc.measure(5,1)
qc.measure(8,2)
qc.measure(11,3)
qc.measure(12,4)
```

```python
qc.draw(output='mpl')
```

```python
state_sim = Aer.get_backend("qasm_simulator")
job = execute(qc, state_sim)
counts = job.result().get_counts(qc)

print(counts)
```

```python
plot_histogram(counts)
```

On entangling the qubits 00, we get + state, i.e, 50% of time we are adding the numbers "0000" and "1110" (thus getting "01110") and 50% of time, we are adding "0001" and "1111" (thus getting "10000").
