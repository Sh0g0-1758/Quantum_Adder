The code below shows the implementaion of a single full adder circuit.

```python
# importing required packages
from qiskit import *
from qiskit import QuantumCircuit
from qiskit.tools.visualization import plot_histogram
from pylatexenc import *
```

```python
# creating a qunatum circuit of 4 qubits and 2 classical bits
qc = QuantumCircuit(4,2)
qc.x(2)
qc.x(0)
```

```python
# the following is the logic of the full adder
qc.ccx(1,2,3)
```

```python
qc.cx(1,2)
```

```python
qc.ccx(0,2,3)
```

```python
qc.cx(0,2)
```

```python
# mapping the value of qubit to a particular classical bit
qc.measure(2,0)
```

```python
qc.measure(3,1)
```

```python
# drawing the circuit
qc.draw(output='mpl')
```

```python
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

```python
# here we test the full adder thus created using the values set earlier.
state_sim = Aer.get_backend("qasm_simulator")
job = execute(qc, state_sim)
counts = job.result().get_counts(qc)

print(counts)
```

```python
plot_histogram(counts)
```
