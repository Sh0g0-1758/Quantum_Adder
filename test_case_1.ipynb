The following code is implementing the full adder created for 4 bit numbers. 

```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
from pylatexenc import *
from qiskit_aer import AerSimulator
```

```python
# quantum circuit of 13 qubits and 5 classical bits
qc = QuantumCircuit(13,5)
```

```python
# setting the values according to the test case 1
qc.x(4)
qc.x(2)
qc.x(5)
qc.x(11)
qc.barrier()
```

```python
# first full adder
qc.ccx(1,2,3)
qc.cx(1,2)
qc.ccx(0,2,3)
qc.cx(0,2)
qc.barrier()
```

```python
# second full adder
qc.ccx(4,5,6)
qc.cx(4,5)
qc.ccx(3,5,6)
qc.cx(3,5)
qc.barrier()
```

```python
# third full adder
qc.ccx(7,8,9)
qc.cx(7,8)
qc.ccx(6,8,9)
qc.cx(6,8)
qc.barrier()
```

```python
# fourth full adder
qc.ccx(10,11,12)
qc.cx(10,11)
qc.ccx(9,11,12)
qc.cx(9,11)
qc.barrier()
```

```python
# measuring the values of the qubits. 
qc.measure(2,0)
qc.measure(5,1)
qc.measure(8,2)
qc.measure(11,3)
qc.measure(12,4)
```

```python
qc.draw(output='mpl')
```

```python
# simulating using qasm_simulator
state_sim = Aer.get_backend("qasm_simulator")
job = execute(qc, state_sim)
counts = job.result().get_counts(qc)

print(counts)
```

```python
plot_histogram(counts)
```
